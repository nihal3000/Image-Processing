```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\DS17\Desktop\IVP 011\Spatial filtering techniques\download.jpeg", cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Gaussian filter using opencv
gauss_img = cv2.GaussianBlur(image, (5,5), 0) # kernel_size=(5,5), SigmaX=0(default)
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.title("Original image"), plt.xticks([]), plt.yticks([])
plt.imshow(image)

plt.subplot(1,2,2)
plt.title("Gaussian filtered image"), plt.xticks([]), plt.yticks([])
plt.imshow(gauss_img)
plt.show()
```

```python
# Sobel filter
sobel_X = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_Y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Now combining the two gradients to get the magnitude of the image
img_mag = cv2.magnitude(sobel_X, sobel_Y)

plt.figure(figsize=(10,6))
plt.subplot(1,4,1)
plt.title("Original image"), plt.xticks([]), plt.yticks([])
plt.imshow(image)

plt.subplot(1,4,2)
plt.title("Sobel_X filtered image"), plt.xticks([]), plt.yticks([])
plt.imshow(sobel_X)

plt.subplot(1,4,3)
plt.title("Sobel_Y filtered image"), plt.xticks([]), plt.yticks([])
plt.imshow(sobel_Y)


plt.subplot(1,4,4)
plt.title("Magnitude image"), plt.xticks([]), plt.yticks([])
plt.imshow(img_mag)
plt.show()
```

```python
# Laplacian filter
lap_img = cv2.Laplacian(image, cv2.CV_64F, ksize=3)

plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.title("Original image"), plt.xticks([]), plt.yticks([])
plt.imshow(image)

plt.subplot(1,2,2)
plt.title("Gaussian filtered image"), plt.xticks([]), plt.yticks([])
plt.imshow(gauss_img)
plt.show()
```
