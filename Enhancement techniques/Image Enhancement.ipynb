```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\sdrfwe\Desktop\tiger.jfif', cv2.IMREAD_GRAYSCALE)

# Define the min and max values for stretching
min_val = np.min(image)
max_val = np.max(image)

# Apply contrast stretching
stretched = (image - min_val) * (255 / (max_val - min_val))
stretched = stretched.astype(np.uint8)

# Display the original and stretched images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Contrast Stretched Image')
plt.imshow(stretched, cmap='gray')
plt.show()
```

```python
# Log anfd inverse Log
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(r'C:\Users\sdrfwe\Desktop\tiger.jfif', cv2.IMREAD_GRAYSCALE)

# Perform logarithmic transformation
c = 255 / np.log(1 + np.max(image))
log_transformed = c * np.log(1 + image)

# Perform inverse logarithmic transformation
inverse_transformed = np.exp(log_transformed / c) - 1
inverse_transformed = np.clip(inverse_transformed, 0, 255).astype(np.uint8)

# Display the original, log transformed, and inverse transformed images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Log Transformed Image')
plt.imshow(log_transformed.astype(np.uint8), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Inverse Log Transformed Image')
plt.imshow(inverse_transformed, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# Power law Transformations
image = cv2.imread(r'C:\Users\sdrfwe\Desktop\tiger.jfif', cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gamma = 1.7
c = 1
power_trans = c * (image/255.0)**gamma
power_trans = np.clip(power_trans*255, 0, 255).astype(np.uint8)

plt.figure(figsize=(8,6))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(power_trans)
plt.title("Transformed image")
plt.show()
```

```python
# Power law Transformations (nth root)
image = cv2.imread(r'C:\Users\sdrfwe\Desktop\tiger.jfif', cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

n = 0.7
c = 1
power_trans = c * (image/255.0)**(1/n)
power_trans = np.clip(power_trans*255, 0, 255).astype(np.uint8)

plt.figure(figsize=(8,6))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(power_trans)
plt.title("Transformed image")
plt.show()
```
