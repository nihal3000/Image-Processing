Types of Image Segmentation:
1. Thresholding
2. Region-Based
3. Edge-Based
4. Semantic
5. Instance

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Thresholding
image = cv2.imread(r"C:\Users\DS17\Downloads\med.webp",0)

threshold_value = 128

_, thresh_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.imshow(thresh_image, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Region-Based
image = cv2.imread(r"C:\Users\DS17\Downloads\flower.webp")

gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Reducing the noise
blurred_img = cv2.GaussianBlur(gray_img, (5,5), 0)

# Binary image
_, bin_img = cv2.threshold(blurred_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Finding contours
contours, _ = cv2.findContours(bin_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw the contours to the original image
cv2.drawContours(image, contours, -1, (0,255,0), 2)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,7))
plt.xticks([]), plt.yticks([])
plt.imshow(image)
plt.show()
```

```python
# Edge-Based
image = cv2.imread(r"C:\Users\DS17\Downloads\med.webp",0)

edges = cv2.Canny(image, 100, 200)

plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.xticks([]), plt.yticks([])
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.xticks([]), plt.yticks([])
plt.imshow(edges, cmap='gray')
plt.show()
```
