```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Adding gaussian noise to an image
image = cv2.imread(r"C:\Users\DS17\Downloads\nature.jpeg", cv2.IMREAD_GRAYSCALE)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
image.shape
```

```python
mean = 0
sigma = 25
gauss_noise = np.random.normal(mean, sigma, image.shape)
noisy_image = np.clip(image + gauss_noise, 0, 255).astype(np.uint8)
plt.figure(figsize=(8,6))
plt.title("Image with gaussian noise"), plt.xticks([]), plt.yticks([])
plt.imshow(noisy_image, cmap='gray')
plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
image.shape
```

```python
# Rayleigh noise
sigma = 25
rayleigh_noise = np.random.rayleigh(sigma, image.shape)
rayleigh_noisy_image = np.clip(image + rayleigh_noise, 0, 255).astype(np.uint8)
plt.figure(figsize=(8,6))
plt.title("Image with rayleigh noise"), plt.xticks([]), plt.yticks([])
plt.imshow(rayleigh_noisy_image, cmap='gray')
plt.show()
```

```python
noisy_image = cv2.cvtColor(noisy_image, cv2.COLOR_RGB2GRAY)
noisy_image.shape
```

```python
plt.subplot(1, 2, 1)
plt.hist(rayleigh_noisy_image, bins=50, alpha=0.7, edgecolor='black')
plt.title('Histogram of Rayleigh Noise')
plt.xlabel('Noise Value')
plt.ylabel('Frequency')

# Plot a sample of the Rayleigh noise
plt.subplot(1, 2, 2)
plt.hist(noisy_image, bins=50, alpha=0.7, edgecolor='black')
plt.title('Histogram of Gaussian Noise')
plt.xlabel('Noise Value')
plt.ylabel('Frequency')
```

```python
# Erlang noise
b = 3
a = 2


erlang_noise = np.random.gamma(b, 1/a, image.shape)

plt.subplot(1, 2, 1)
plt.hist(erlang_noise, bins=50
         , alpha=0.7, edgecolor='black')
plt.title('Histogram of Erlang Noise')
plt.xlabel('Noise Value')
plt.ylabel('Frequency')

# Plot a sample of the Erlang noise
plt.subplot(1, 2, 2)
plt.plot(erlang_noise[:1000], color='purple')
plt.title('Sample of Erlang Noise')
plt.xlabel('Sample Index')
plt.ylabel('Noise Value')

plt.tight_layout()
plt.show()
```

```python
erlang_image = np.clip(image + erlang_noise, 0, 255).astype(np.uint8)

plt.hist(erlang_image, bins=50, alpha=0.7, edgecolor='black')
plt.title('Histogram of erlang Noise')
plt.xlabel('Noise Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Exponential noise
lambda_ = 1 # rate parameter
expo_noise = np.random.exponential(scale=1/lambda_, size=image.shape)

# Plotting the noise
plt.figure(figsize=(8,6))
plt.hist(expo_noise, bins=50
         , alpha=0.7, edgecolor='black')
plt.title("Exponential noise")
plt.xlabel("Noise value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Adding exponential noise to the image
expo_image = np.clip(image + expo_noise, 0, 255).astype(np.uint8)

# plotting the noise image
plt.figure(figsize=(8,6))
plt.imshow(expo_image, cmap='gray')
plt.title("Exponential noisy image"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Uniform noise
a,b = -1, 1 # uniform noise between -1 to 1
uniform_noise = np.random.uniform(low=a, high=b, size=image.shape)
# Plotting the noise
plt.figure(figsize=(6,6))
plt.hist(uniform_noise, bins=50
         , alpha=0.7, edgecolor='black')
plt.title("Uniform noise")
plt.xlabel("Noise value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Image with uniform noise
uniform_image = np.clip(image + uniform_noise, 0, 255).astype(np.uint8)
# plotting the noise image
plt.figure(figsize=(6,6))
plt.imshow(uniform_image, cmap='gray')
plt.title("Uniform noise image"), plt.xticks([]), plt.yticks([])
plt.show()
```
