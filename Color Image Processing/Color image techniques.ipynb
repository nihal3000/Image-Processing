```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\DS17\Desktop\IVP 011\Color Image Processing\download.jpeg")
plt.imshow(image)
```

```python
color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(color_image)
```

```python
# RGB to Grascale conversion
gray_image = cv2.cvtColor(color_image, cv2.COLOR_RGB2GRAY)
plt.title("Grayscale image"), plt.xticks([]), plt.yticks([])
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
# RGB to HSV conversion
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
hsv_image = cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB)
plt.title("HSV image"), plt.xticks([]), plt.yticks([])
plt.imshow(hsv_image)
plt.show()
```
